```python
#Min Max Scaling
##MinMaxScaler new_X = (x - min(X)) / (max(X) - min(X))
from sklearn import preprocessing
plt.rcParams['figure.figsize'] = [40, 40]
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
newDf = min_max_scaler.fit_transform(df)

newDf = pd.DataFrame(newDf)

for i in range(18):
    m = np.mean(newDf[i])
    cols = np.where((newDf[i] >= m - (0.8 * m)) & (newDf[i] <= m + (0.8 * m)), 'b', 'r')
    plt.subplot(4,5,i + 1)
    plt.scatter(newDf.index,newDf[i], c = cols)
plt.show()
```

```python
#StandardScaler / ZScore
#Standard Scaler x' = (x - mean(X)) / std(X)

z_scaler = preprocessing.StandardScaler()

newDataframe = z_scaler.fit_transform(df)

newDataframe = pd.DataFrame(newDataframe)

for i in range(18):
    m = np.mean(newDataframe[i])
    cols = np.where((newDataframe[i] >= -0.8) & (newDataframe[i] <= 0.8), 'y', 'r')
    plt.subplot(4,5,i + 1)
    plt.scatter(newDataframe.index,newDataframe[i], c = cols)
plt.show()
```

```python
#Max Absolute
#MaxAbsoluteScaler x' = x / max(abs(X))

z_scaler = preprocessing.MaxAbsScaler()
newDataframe = z_scaler.fit_transform(df)

newDataframe = pd.DataFrame(newDataframe)

for i in range(18):
    m = np.mean(newDataframe[i])
    cols = np.where((newDataframe[i] >= m - (0.8 * m)) & (newDataframe[i] <= m + (0.8 * m)), 'g', 'r')
    plt.subplot(4,5,i + 1)
    plt.scatter(newDataframe.index,newDataframe[i], c = cols)


plt.show()
```
